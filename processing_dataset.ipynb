```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from urllib.parse import urlparse
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("dataset/new_malicious.csv")
```

```python
df["type"]
```

```python
df.columns
```

```python
df.drop_duplicates(inplace=True)
```

```python

train_df, test_df = train_test_split(
    df,
    test_size=0.2,
    random_state=42,
    stratify=df["label"]
)
```

```python
#test_df için

column_names = ['@', '?', '-', '=', '.', '#', '%','+', '$', '!', '*', ',', '//']
test_df["special_characters"] = test_df[column_names].sum(axis=1)
```

```python
train_df.drop(columns=column_names, axis=1, inplace=True)
test_df.drop(columns=column_names, axis=1, inplace=True)
```

```python
train_df.drop(columns=["type"], axis=1, inplace=True)
test_df.drop(columns=["type"], axis=1, inplace=True)
```

```python
def get_path_deep(url):
    path_deep = 0
    parsed = urlparse(url)
    path = parsed.path
    path_deep = path.count('/')
    return path_deep

def is_path_deep(path_deep):
    return 1 if path_deep > 3 else 0

def get_path_underscore(url):
    path_underscore = 0
    parsed = urlparse(url)
    path = parsed.path
    path_underscore = path.count("_")
    return path_underscore

def has_index_php(url):
    parsed = urlparse(url)
    path = parsed.path
    return 1 if 'index.php' in path.lower() else 0

def has_option_param(url):
    return 1 if 'option' in url.lower() else 0
    
```

```python
#train_df için

train_df["path_deep"] = train_df["url"].apply(lambda x : get_path_deep(x))
train_df["is_path_deep"] = train_df["path_deep"].apply(lambda x : is_path_deep(x))
train_df["get_path_underscore_count"] = train_df["url"].apply(lambda x : get_path_underscore(x))
train_df["has_index_php"] = train_df["url"].apply(lambda x : has_index_php(x))
train_df["has_option_param"] = train_df["url"].apply(lambda x : has_option_param(x))
```

```python
#test_df için

test_df["path_deep"] = test_df["url"].apply(lambda x : get_path_deep(x))
test_df["is_path_deep"] = test_df["path_deep"].apply(lambda x : is_path_deep(x))
test_df["get_path_underscore_count"] = test_df["url"].apply(lambda x : get_path_underscore(x))
test_df["has_index_php"] = test_df["url"].apply(lambda x : has_index_php(x))
test_df["has_option_param"] = test_df["url"].apply(lambda x : has_option_param(x))
```

```python
train_df.drop(columns="url",axis=1,inplace=True)
test_df.drop(columns="url",axis=1,inplace=True)
```

```python
train_df.drop_duplicates(inplace=True)
test_df.drop_duplicates(inplace=True)
```

```python
# divide scan_date to year, month, day ... 

def scan_date_year(scan_date):
    year = scan_date[:4]
    year_int = int(year)
    return year_int
def scan_date_month(scan_date):
    month = scan_date[5:7]
    month_int = int(month)
    return month_int
def scan_date_day(scan_date):
    day = scan_date[8:10]
    day_int = int(day)
    return day_int
def scan_date_hour(scan_date):
    hour = scan_date[11:13]
    hour_int = int(hour)
    return hour_int
def scan_date_minute(scan_date):
    minute = scan_date[14:16]
    minute_int = int(minute)
    return minute_int
def scan_date_second(scan_date):
    second = scan_date[17:19]
    second_int = int(second)
    return second_int
```

```python


#train_df

train_df['year'] = train_df['scan_date'].apply(lambda x : (scan_date_year(x)))
train_df['month'] = train_df['scan_date'].apply(lambda x : (scan_date_month(x)))
train_df['day'] = train_df['scan_date'].apply(lambda x : (scan_date_day(x)))
train_df['hour'] = train_df['scan_date'].apply(lambda x : (scan_date_hour(x)))
train_df['minute'] = train_df['scan_date'].apply(lambda x : (scan_date_minute(x)))
train_df['second'] = train_df['scan_date'].apply(lambda x : (scan_date_second(x)))
```

```python

#test_df

test_df['year'] = test_df['scan_date'].apply(lambda x : (scan_date_year(x)))
test_df['month'] = test_df['scan_date'].apply(lambda x : (scan_date_month(x)))
test_df['day'] = test_df['scan_date'].apply(lambda x : (scan_date_day(x)))
test_df['hour'] = test_df['scan_date'].apply(lambda x : (scan_date_hour(x)))
test_df['minute'] = test_df['scan_date'].apply(lambda x : (scan_date_minute(x)))
test_df['second'] = test_df['scan_date'].apply(lambda x : (scan_date_second(x)))
```

```python

train_df.drop(columns="scan_date", axis=1, inplace =True)
test_df.drop(columns="scan_date", axis=1, inplace =True)
```

```python
train_df['hour_normalized'] = (train_df['hour'] - train_df['hour'].min()) / (train_df['hour'].max() - train_df['hour'].min())
train_df['minute_normalized'] = (train_df['minute'] - train_df['minute'].min()) / (train_df['minute'].max() - train_df['minute'].min())
train_df['second_normalized'] = (train_df['second'] - train_df['second'].min()) / (train_df['second'].max() - train_df['second'].min())
```

```python
test_df['hour_normalized'] = (test_df['hour'] - test_df['hour'].min()) / (test_df['hour'].max() - test_df['hour'].min())
test_df['minute_normalized'] = (test_df['minute'] - test_df['minute'].min()) / (test_df['minute'].max() - test_df['minute'].min())
test_df['second_normalized'] = (test_df['second'] - test_df['second'].min()) / (test_df['second'].max() - test_df['second'].min())
```
